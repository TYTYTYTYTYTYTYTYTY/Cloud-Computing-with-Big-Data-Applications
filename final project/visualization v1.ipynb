```python
import re
pd.set_option("display.precision", 10)

re1='.*?'	# Non-greedy match on filler
re2='(\\d+)'	# Integer Number 1
re3='.*?'	# Non-greedy match on filler
re4='([+-]?\\d*\\.\\d+)(?![-+0-9\\.])'	# Float 1
re5='.*?'	# Non-greedy match on filler
re6='([+-]?\\d*\\.\\d+)(?![-+0-9\\.])'	# Float 2

filenames = ["part-00000 (4)","part-00000 (5)","part-00001", "part-00001 (1)"]
filepath = "/Users/muzhou/Desktop/CS/427/Cloud-Computing-with-Big-Data-Applications/final project/data/playground-data1/"
df = pd.DataFrame(columns = ["long", "lat", "cluster"])

with open(filepath+filenames[0]) as fp:
    lines = list(fp)
    cnt =0
    for line in lines:
        if cnt >10000:
            break
        rg = re.compile(re1+re2+re3+re4+re5+re6,re.IGNORECASE|re.DOTALL)
        m = rg.search(line)
        if m:
            int1=m.group(1)
            float1=m.group(2)
            float2=m.group(3)
            #print((float1, float2,int1))
            df = df.append(pd.DataFrame({"long":[float1] , "lat":[float2], "cluster":[int1]}))

        cnt+=1
        
print(df.head(5))

```

```python
df1 = df.sample(n= 200, random_state=12345)

incidents = folium.map.FeatureGroup()


for index, row in df1.iterrows():
    lat = row[0]
    lng = row[1]
    cluster = int(row[2])
    colors = ['pink', 'gray', 'darkblue', 'beige','red']
    incidents.add_child(
        folium.Marker(
            [lat, lng],
            icon= folium.Icon(color = colors[cluster], icon = str(cluster))
        )
    )
    

```
